```python
import numpy as np
import matplotlib.pyplot as plt 

e = 1.6e-19
k = 1.38e-23
u = 1.6e-19

E = np.linspace(-0.5,0.5,1001)

def Fn(T):
    return 1/(1+np.exp(((E-u)*e)/(k*T)))

plt.plot(E,Fn(100), label ='T = 100k')
plt.plot(E,Fn(300), label ='T = 300k')
plt.plot(E,Fn(500), label ='T = 500k')
plt.plot(E,Fn(700), label ='T = 700k')
plt.plot(E,Fn(900), label ='T = 900k')
plt.plot(E,Fn(50), label ='T = 50k')
plt.legend(loc="best", prop={'size':12})
plt.ylim(-0,1)
plt.xlabel("Energy eV")
plt.ylabel("F(e)")
plt.title("Fermi-Dirac Distribution for U = 0")
plt.show()
```
